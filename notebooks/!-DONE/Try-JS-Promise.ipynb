# JavaScript Promise 简洁教程

## 一、Promise 是什么？

Promise 是 JavaScript 中处理异步操作的对象，用于避免"回调地狱"（多层嵌套回调）的问题。它代表一个"承诺"——无论成功或失败，都会给出答复。

### Promise 的三种状态
- **Pending（进行中）**：初始状态
- **Resolved/Fulfilled（已完成）**：异步操作成功
- **Rejected（已失败）**：异步操作失败

**重要特点**：状态改变是单向的，一旦改变就固定，不会再变。



## 二、基本用法

### 创建 Promise

```javascript
const promise = new Promise((resolve, reject) => {
  // 异步操作
  if (/* 成功条件 */) {
    resolve('成功的结果');  // 改变状态为 resolved
  } else {
    reject('失败的原因');   // 改变状态为 rejected
  }
});
```

### then() 和 catch()

```javascript
promise
  .then(result => {
    console.log('成功:', result);
  })
  .catch(error => {
    console.log('失败:', error);
  });
```

**关键点**：
- `then()` 返回新的 Promise，可以链式调用
- 如果链中某个 `then()` 抛出错误，会被后续的 `catch()` 捕获
- `catch()` 捕获错误后，可以继续 `then()`

### 完整示例：异步延时操作

```javascript
function delay(ms, value) {
  return new Promise((resolve) => {
    setTimeout(() => resolve(value), ms);
  });
}

delay(1000, '第一步完成')
  .then(result => {
    console.log(result);
    return delay(1000, '第二步完成');
  })
  .then(result => {
    console.log(result);
    return delay(1000, '第三步完成');
  })
  .then(result => {
    console.log(result);
  });
```



## 三、Promise 静态方法

### Promise.resolve(value)
快速创建已解决的 Promise

```javascript
Promise.resolve(42).then(value => {
  console.log(value);  // 42
});
```

### Promise.reject(error)
快速创建已拒绝的 Promise

```javascript
Promise.reject('出错了').catch(error => {
  console.log(error);  // 出错了
});
```

### Promise.all([p1, p2, ...])
并行执行多个 Promise，**所有都成功**才算成功（&&）

```javascript
Promise.all([
  fetch('/api/user'),
  fetch('/api/posts')
]).then(results => {
  console.log('所有请求完成', results);
}).catch(error => {
  console.log('任何一个失败就执行此处', error);
});
```

### Promise.race([p1, p2, ...])
**只要任何一个**Promise 改变状态就结束（||）

```javascript
Promise.race([
  fetch('/api/primary'),
  fetch('/api/backup')
]).then(result => {
  console.log('获得最快的响应', result);
});
```



## 四、实际应用：用 Promise 包装 AJAX

```javascript
function get(url) {
  return new Promise((resolve, reject) => {
    const xhr = new XMLHttpRequest();
    xhr.open('GET', url, true);
    
    xhr.onload = function() {
      if (xhr.status === 200) {
        resolve(xhr.response);
      } else {
        reject(new Error(`HTTP ${xhr.status}`));
      }
    };
    
    xhr.onerror = () => {
      reject(new Error('网络错误'));
    };
    
    xhr.send();
  });
}

// 使用：
get('/api/data')
  .then(response => console.log('数据:', response))
  .catch(error => console.log('错误:', error));
```

## 五、常见错误与最佳实践

| 场景 | 错误做法 | 正确做法 |
|------|--------|--------|
| 链式调用 | `promise.then(...).catch(...).then(...)` 不执行最后的then | 理解 catch 后仍可继续 then |
| 忘记 return | 在 then 中异步操作但不 return | 每个 then 都应 return Promise |
| 状态固化 | 期望改变已确定的 Promise 状态 | 记住状态改变是单向的 |
| 错误处理 | 只用 catch 不用 try-catch | Promise 的 catch 只捕获异步错误 |

## 六、总结

**Promise 的核心优势**：
✅ 解决回调地狱问题  
✅ 让异步代码更易读、易维护  
✅ 支持链式调用  
✅ 便于组合多个异步操作
